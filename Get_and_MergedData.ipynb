```python
df1 = pd.read_csv('Earthquake_Dataset1.csv')
df1
```

```python
df2 = pd.read_csv('Earthquake_Dataset2.csv')
df2
```

```python
# check column
df1.columns
```

```python
# check column
df2.columns
```

```python
# ลบ column ที่ไม่จำเป็นออก
df1_E = df1.drop(columns=['magType', 'rms', 'net', 'id', 'updated', 'type',
                        'horizontalError', 'depthError', 'magError', 'magNst', 'status','locationSource', 'magSource'])
df1_E
```

```python
# ลบ column ที่ไม่จำเป็นออก
df2_E = df2.drop(columns=['magnitude type', 'Unnamed: 7', 'Unnamed: 8'])
df2_E
```

```python
# แก้ไขชื่อ columns ของตาราง data
df1_E.columns = ['date', 'latitude', 'longitude', 'depth', 'magnitude', 'nst','gap', 'dmin', 'place']
df1_E
```

```python
merged_df = pd.concat([df1_E, df2_E], ignore_index=True)
merged_df
```

```python
merged_df.to_csv('Merged_EarthQuake_data.csv', index=False)
```
